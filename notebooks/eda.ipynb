```python
import pandas as pd

df = pd.read_csv("../data/raw/bank_transactions.csv")

print(df.head())
```

```python
print(df.columns)
```

```python
print(df.info())
print(df.isnull().sum())
```

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'] ,format="%d-%m-%y")
df['Transaction_Amount'] = df['Transaction_Amount'].astype(float)
df['Is_Fraud'] = df['Is_Fraud'].astype(int)
```

```python
print(df['Transaction_Date'].head())
```

```python
df['Transaction_DateTime'] = pd.to_datetime(
    df['Transaction_Date'].astype(str) + ' ' + df['Transaction_Time']
)
```

```python
print(df['Transaction_DateTime'].head())
```

```python
df.to_csv("../data/processed/cleaned_transactions.csv", index=False)
```
